```python
import numpy as np
def newtonSqrt(num, Iteration):
    approx = 0.5 * num
    for i in range(Iteration):
        betterapprox = 0.5 * (approx + num/approx)
        approx = betterapprox
    return betterapprox

print(newtonSqrt(100, 10))
print(newtonSqrt(4, 10))
print(newtonSqrt(1, 10))
```
