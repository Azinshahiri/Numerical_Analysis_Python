```python
import numpy as np
import matplotlib.pyplot as plt
a= float(input(" please put a value:"))
n= int(input("please enter the number of iterations:"))
x= float(input("The start value: "))

print("value:", a)
print("iterations:", n)
print("The start value:", x)
for i in range(n):
   y=x*x-a
   y_prime=2*x
   x=x- y/y_prime
   print("iterations", i+1, ":", x)

# print("exact value:", a**0.5)
# print("exact value is:", a**0.5)
# print("error:", abs(a**0.5 -x))

plt.grid()
m=np.linspace(-9,9,100)
y=m*m-a
plt.plot(m,y)
plt.plot(x,0, marker="o",color= "red")





```
